```python
import pandas as pd
import numpy as np

#read data
df = pd.read_csv('csv/cleaned_loged.csv', low_memory=False)

#pivot table
df = pd.pivot_table(data=df,index=['src_ip'],values=['byte'],aggfunc={np.sum})
df.to_csv("csv/sip_bytes.csv")

#initalize data
df = pd.read_csv("csv/sip_bytes.csv",skiprows=3,sep=',',names=['src_ip','bytes'])
df['data'] = "'" + df['src_ip'].astype(str) + " - " + df['bytes'].astype(str) + "'"

#percentage calculation
df['percentage']=(df['bytes']/df['bytes'].sum())*100
print(df['percentage'].sum())

#sort values by percentage in decending order
df=df.sort_values(by = 'percentage', ascending = False)

#stored data
df.to_csv("csv/sip_bytes.csv",index=False)
df
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('csv/cleaned_loged.csv', low_memory=False)

df = pd.pivot_table(data=df,index=['src_ip'],values=['byte','second'],aggfunc={len,np.sum})
df.to_csv("csv/sip_bytes_connections_durations.csv")

df = pd.read_csv("csv/sip_bytes_connections_durations.csv",skiprows=3,sep=',',names=['src_ip','connections','bytes','d_con','durations'])
df = df.loc[:, ['src_ip','connections','bytes','durations']]

#stored data
df.to_csv("csv/sip_bytes_connections_durations.csv",index=False)
df
```

```python
sip_connections = df
sip_connections['data'] = "'" + sip_connections['src_ip'].astype(str) + " - " + sip_connections['connections'].astype(str) + "'"
sip_connections['percentage'] = (sip_connections['connections']/sip_connections['connections'].sum())*100
sip_connections = sip_connections.loc[:, ['src_ip','connections','data','percentage']]
sip_connections = sip_connections.sort_values(by = 'percentage', ascending = False)
sip_connections.to_csv("csv/sip_connections.csv",index=False)
sip_connections
```

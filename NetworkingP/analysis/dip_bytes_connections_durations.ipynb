```python
import pandas as pd
import numpy as np

#read data
df = pd.read_csv('csv/cleaned_loged.csv', low_memory=False)

#pivot table
df = pd.pivot_table(data=df,index=['dst_ip'],values=['byte'],aggfunc={np.sum})
df.to_csv("csv/dip_bytes.csv")

#initalize data
df = pd.read_csv("csv/dip_bytes.csv",skiprows=3,sep=',',names=['dst_ip','bytes'])
df['data'] = "'" + df['dst_ip'].astype(str) + " - " + df['bytes'].astype(str) + "'"

#percentage calculation
df['percentage']=(df['bytes']/df['bytes'].sum())*100
print(df['percentage'].sum())

#sort values by percentage in decending order
df=df.sort_values(by = 'percentage', ascending = False)

#stored data
df.to_csv("csv/dip_bytes.csv",index=False)
df
```
